```python
import pandas as pd
import matplotlib.pyplot as plt
d = {'Movie': ["Star Wars: The Last Jedi", 
             "Jumanji: Welcome to the Jungle",
             "Pitch Perfect 3",
             "The Greatest Showman",
             "Ferdinand"], 
     'Weekend Gross': [71565498, 36169328,19928525,8805843,7316746]
    }
df = pd.DataFrame(data=d)
#df.columns = ["movie","Weekend Gross"]

fig, ax = plt.subplots(figsize=(15,5))
plt.bar(df["Movie"],df["Weekend Gross"]/1000000,width=0.5)
plt.ylabel("Weekend Gross (in million)",fontsize=16)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,5))
plt.barh(df["Movie"],df["Weekend Gross"]/1000000,hatch='/')
plt.xlabel("Weekend Gross (in million)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
d = {'Movie': ["Star Wars: The Last Jedi", 
             "Jumanji: Welcome to the Jungle",
             "Pitch Perfect 3",
             "The Greatest Showman",
             "Ferdinand"], 
     'Weekend Gross': [71565498, 36169328,19928525,8805843,7316746]
    }
df = pd.DataFrame(data=d)
#df.columns = ["movie","Weekend Gross"]

fig, ax = plt.subplots(figsize=(15,5))
plt.plot(range(len(df['Movie'])),df["Weekend Gross"]/1000000, '*')
plt.ylabel("Weekend Gross (in million)",fontsize=16)
ax.set_xticks(range(len(df['Movie'])))
ax.set_xticklabels(df['Movie'],fontsize=12)
plt.show()
```
