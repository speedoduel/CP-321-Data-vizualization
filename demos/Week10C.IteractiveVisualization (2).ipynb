```python
import simpleguitk
from pandas import read_csv

## global variable
width = 1000
height = 600
edge = 80
series = read_csv('NASDAQComposite.csv', header=0, index_col=0)
series = series.dropna()
startDate = series.index[0]
endDate = series.index[len(series)-1]
colPlot = 'Close'
message = 'NASDAQ Composite Close'

dayInfoPos = [edge,edge]
date = ''
dayInfo = ''

## define helpers

def drawAxis(canvas):
    global ox, oy, axisWidth, axisHeight
    ox = edge
    oy = height-edge
    global axisWidth
    axisWidth = width - 2*edge
    axisHeight = height - 2*edge
    canvas.draw_polygon([(edge, edge), (width-edge, edge), (width-edge, height-edge), (edge, height-edge)], 1,'Black', 'White')
    
def drawLine(canvas, x, y, lineWidth = 1, color = 'Red'):
    if len(x)>0 and len(x) == len(y):
        pointStart = [ox, oy-float(axisHeight)*(y.iloc[0]-y.min())/(y.max()-y.min())]
        dx = float(axisWidth)/len(x)
        for i in range(len(x)):
            dy = float(axisHeight)*(y.iloc[i]-y.min())/(y.max()-y.min())
            pointEnd = [pointStart[0] + dx, oy-dy]
            canvas.draw_line(pointStart,pointEnd,lineWidth,color)
            pointStart = pointEnd
        ## add x and y labels
        for i in range(0,5):
            canvas.draw_line([ox, oy-i*float(axisHeight)/4],[ox+5, oy-i*float(axisHeight)/4],lineWidth,'Black')
            canvas.draw_text(str(int(i*(y.max()-y.min())/4+y.min())), [ox-70, oy-i*float(axisHeight)/4], 10, 'Black')
        for i in range(round(len(x)/30)+1):
            canvas.draw_line([ox+dx*i*30,oy],[ox+dx*i*30, oy-5],lineWidth,'Black')
            canvas.draw_text(x[i*30], [ox+dx*i*30-10, oy+20], 10, 'Black') 

## Event handlers
def click_button_volume():
    global colPlot, message
    colPlot = 'Volume'
    message = 'NASDAQ Composite Volume'
    
def click_button_close():
    global colPlot, message
    colPlot = 'Close'
    message = 'NASDAQ Composite Close'

def draw(canvas):
    drawAxis(canvas)
    drawLine(canvas, series.loc[startDate:endDate].index, series.loc[startDate:endDate][colPlot])
    canvas.draw_text(message, [width/4, 40], 30, 'Red')
    canvas.draw_text(dayInfo, dayInfoPos, 10, 'Blue')

def enter_start(input):
    global startDate
    startDate = input

def enter_end(input):
    global endDate
    endDate = input    

def click_mouse(pos):
    global date, dayInfo, dayInfoPos
    dayInfoPos = pos
    date = series.loc[startDate:endDate].index[round(len(series.loc[startDate:endDate])
                                                     *float(dayInfoPos[0]-ox)/axisWidth)]
    dayInfo = date + '\n' + series.loc[date].to_string()
    
## Create a frame
frame = simpleguitk.create_frame('DataVisualization',width, height)

## Register event handler
frame.set_draw_handler(draw)
frame.set_canvas_background("White")
frame.add_button('Visualize Volume',click_button_volume)
frame.add_button('Visualize Close Price',click_button_close)
frame.add_input('Enter Start Date', enter_start, 100)
frame.add_input('Enter End Date', enter_end, 100)
frame.set_mouseclick_handler(click_mouse)

## Start frame and timers
frame.start()
```
