```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] =[8,6]
t = np.linspace(0.0, 2.0, 21)
s = np.sin(2 * np.pi * t)

fig, ax = plt.subplots(facecolor=(0.3,0.5,0.8))
ax.set_facecolor('#eafff5')
ax.set_title('Sin Function',color='0.9', size='30')
ax.set_xlabel('time',color='y', size ='20')
ax.set_ylabel('sin',color='xkcd:crimson', size ='20')

ax.plot(t,s, '-^',color='C4', linewidth=1)
ax.set_xticks([0,1,2,3])
ax.set_xticklabels(['year1','year2','year3','year4'])
ax.tick_params(labelcolor='tab:red', labelsize='15')
plt.show()

```
